#### Función de transformación para columnas categóricas

```python
def limpieza(columna):
    
    try: 
        columna = columna.str.strip()
        columna = columna.str.lower()
        columna= columna.str.replace(',','.', regex = False)
        columna = columna.str.replace('$','', regex = False)
        columna = columna.str.replace('-','_', regex = False)
        columna = columna.str.replace(' ','_', regex = False)
    
        return columna
    
    except:
        print(f'Error en la columna{columna}')
        return columna
```

```python
df[columnas_categ] = df[columnas_categ].apply(limpieza)
```

```python
for col in columnas_categ:
    print(f"Valores únicos en '{col}':")
    print(df[col].unique(), "\n")
```

📌 ¿Por qué NaN aparece en float, pero < NA > en int?

1️⃣ En columnas float64, Pandas usa NaN, que es un valor especial de numpy (np.nan).

2️⃣ En columnas int64, no pueden existir nulos (los enteros clásicos no admiten NaN).

3️⃣ En columnas Int64 (nullable), Pandas usa < NA >, que es su propia representación de valores
