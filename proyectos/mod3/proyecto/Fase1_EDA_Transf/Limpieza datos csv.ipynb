```python
# FUNCION PARA LIMPIAR COLUMNAS OBJECT Y CONVERTIR A NUMERICAS:

def clean_numeric_column(df, column):  # df:dataframe, column: el nombre de la columna a limpiar.
    # Paso 1: convierto a texto y limpio s√≠mbolo
    df[column] = df[column].astype(str) # Convierto toda la columna a str, porque .str.replace() solo funciona con texto.
    df[column] = df[column].str.replace('$', '')
    df[column] = df[column].str.replace(',', '.')

    # Paso 2: convertir a float
    # errors = coerce -> Estoy forzando ->‚ÄúIntenta convertir todo a n√∫mero. Si encuentras algo que no se puede convertir (como texto raro o vac√≠o), en vez de dar error, pon un NaN‚Äù.
    df[column] = pd.to_numeric(df[column], errors='coerce') # pd.to_numerics es una funcion de Pandas que  convierte strings en numeros int/float

    # Paso 3: convertir NaN a None para SQL
    df[column] = df[column].where(pd.notnull(df[column]), None) # si el valor es NaN, cambialo a None, sino dejalo tal cual.

    return df  # devuelvo el df modificado.
```

```python
# creo lista con las columnas a las que le voy a aplicar la funcion que he creado.

columnas_a_limpiar = ['monthlyincome', 'monthlyrate']
```

```python
# hago un bucle for para aplicar la funcion a la lista de columnas que he definido.

for col in columnas_a_limpiar:
    clean_numeric_column(df, col)
```

```python
# ya he importado la libreria, hago una funcion para limpiar la columna age:
# necesito pasarlo todo a texto/string primero.

def convertir_edad(valor):  # Definimos la funci√≥n. Le pasaremos una edad, que puede ser: un numero ("42"), un string o un Nulo (NaN)
    if pd.isnull(valor):    # Si el valor es nulo (NaN), lo dejamos como None. 
        return None
    valor = str(valor).strip().lower() # str(valor) lo convierte a texto/string por si viene como numero, con .strip() elimino posibles espacios y .lower()lo paso todo a minusculas.
    try:
        return int(valor) if valor.isdigit() else w2n.word_to_num(valor)  # si el valor es .isdigit() significa que es un numero tipo "42" y uso INT para convertirlo, si no es numero sino letra (forty-two) uso w2n para traducirlo.
    except:
        return None  # Por si hay alg√∫n texto no v√°lido


#con el try me aseguro de cambiar a INT si es un "35" y me de 35 y si esta escrito "twenty" uso w2n para que me de 20 en INT
```

```python
df["age"] = df["age"].apply(convertir_edad)

# con apply aplico la funcion que acabo de hacer a cada fila de la columna AGE para que realice los cambios a INT. 
# El resultado se guarda en la misma columna.
```

```python
def clean_binary_column(df, column):
    # Paso 1: convertir a texto, quitar espacios y pasar a min√∫sculas
    df[column] = df[column].astype(str).str.strip().str.lower()

    # Paso 2: estandarizar valores (s√≠/no)
    def estandarizar(valor):
        if valor in ['yes', '1', 'true']:
            return 'yes'
        elif valor in ['no', '0', 'false']:
            return 'no'
        else:
            return None  # Valor inesperado

    # Aplicamos la estandarizaci√≥n
    df[column] = df[column].apply(estandarizar)
    
    return df
```

```python
for col in columnas_binarias:
    clean_binary_column(df, col)
```

```python
def clean_text_column(df, column):
    df[column] = df[column].astype(str).str.strip().str.lower()
    df[column] = df[column].str.replace('-','_')
    df[column] = df[column].str.replace(' ','_')
    df[column] = df[column].str.replace('marreid','married')
    df[column] = df[column].where(pd.notnull(df[column]), None)
    return df # devuelve el dataframe modificado.

# pd.notnull(df[column]) -> Devuelve True para los valores que NO son nulos, y False para los NaN.
# .where(condici√≥n, valor_si_no_se_cumple)
# Si la condici√≥n es True, deja el valor como est√°.
# Si la condici√≥n es False (es decir, si hay un NaN), lo reemplaza por None.
```
