#### Transformación de los datos : Creamos una funcion para varias columnas

```python
def limpiar_datos(df,columnas):
    
    for columna in columnas:
        
        df[columna]=df[columna].apply(lambda valor:str(valor).replace(' ','_'))
        #quitamos los espacios al ppo y final de cada dato y genermos_ en los espacios entre palabras.
        
    return(df)#aplicamos los resultados directamente en el df
```

```python
columnas_categoricas = df_clientes.select_dtypes(include='O')
columnas_categoricas
```

```python
limpiar_datos(df_clientes,columnas_categoricas)
```
