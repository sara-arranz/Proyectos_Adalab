```python
def artist_getinfo(df_resultados_pop_20):

    artist_info = []
    lista_artistas = [] #para verificar repeticiones

    for element in df_resultados_pop_20 ['artist_name']: 

        if ',' in element.lower(): #cuando hay más de un artista, me quedo con el primero que sale y añado su info
            lista = element.split(',') #pongo todo en minus. Separo por la coma para crear una lista de artistas
            artist = lista[0] #me quedo con el primer artista de la lista (justo lo que hay antes de la coma)

        elif ' ' in element.lower(): #cuando hay un espacio entre dos palabras que pertenecen al nombre del mismo artista
            artist = element.replace(' ', '+') #sustituimos el espacio del nombre compuesto por "+"

        else:
            artist = element.lower()
            

        #llamamos a la API usando el artista que "acabo de depurar":
        
        llamada = requests.get(f"https://ws.audioscrobbler.com/2.0/?method=artist.getinfo&artist={artist}&api_key={"3a3bb82623596aa4da01f7855ba2975b"}&format=json")
        respuesta = llamada.json()

        if 'artist' not in respuesta:  # Evitar error si la API no nos devuelve nada. Ponemos 'artist' en formato string porque es el primer campo que obtenemos de la API en caso de respuesta 200
            continue 
        
        #creo el diccionario para apendearlo a la lista de arriba que tendrá el total de artistas
        artist_ok = artist.replace('+', ' ') #vuelvo a quitar el '+' para poner un espacio para nombres compuestos (ej. J Balvin en lugar de J+Balvin)
        
        if artist_ok.lower() in lista_artistas: #evitar que haya artistas repetidos
            continue

        info = {'artist': artist_ok.lower(),
                'listeners': respuesta['artist']['stats']['listeners'],
                'playcount': respuesta['artist']['stats']['playcount'],
                'bio_content': respuesta['artist']['bio']['content']}
                                
        artist_info.append(info)
        lista_artistas.append(artist_ok.lower())

    return artist_info
```

```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
import time

def get_spotify_top_tracks(year, limit=10):
    """
    Obtiene las canciones más populares de un año desde Spotify Charts,
    recopilando los rankings semanales.

    Parámetros:
    - year (int): Año a consultar (ejemplo: 2021)
    - limit (int): Número de canciones más populares a devolver (por defecto 10)

    Retorna:
    - Lista de las canciones más populares en formato (nombre, artista, conteo).
    """
    track_counter = Counter()

    for month in range(1, 13):  # Iterar de enero (1) a diciembre (12)
        for week in range(1, 5):  # Aproximadamente 4 semanas por mes
            url = f"https://charts.spotify.com/charts/view/regional-global-weekly/{2020}-{month:02d}-{week*7:02d}"
            response = requests.get(url)

            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser')
                tracks = soup.select(".chart-table-track strong")  # Selector de títulos de canciones
                artists = soup.select(".chart-table-track span")  # Selector de artistas

                for track, artist in zip(tracks, artists):
                    track_name = track.text.strip()
                    artist_name = artist.text.replace("by ", "").strip()
                    track_counter[(track_name, artist_name)] += 1  # Contar repeticiones
            else:
                print(f"No se encontró información para {year}-{month}-{week}")

            time.sleep(0.5)  # Pausa para no saturar el servidor

    # Ordenar canciones por cantidad de apariciones y devolver el top
    top_tracks = track_counter.most_common(limit)

    return top_tracks

# Ejemplo de uso
year = 2020
top_tracks = get_spotify_top_tracks(2020)

print(f"Top {len(top_tracks)} canciones de {year}:")
for idx, (track, count) in enumerate(top_tracks, start=1):
    print(f"{idx}. {track[0]} - {track[1]} (Apareció {count} veces)")
```
