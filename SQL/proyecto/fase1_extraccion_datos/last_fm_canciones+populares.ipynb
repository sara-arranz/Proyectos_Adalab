```python
import requests
import pandas as pd

url = 'https://ws.audioscrobbler.com/2.0/?method=chart.gettoptracks&api_key=a0f04126345752a0660a9831f785ce20&format=json'

# Definimos los años de inicio y fin
from_year = int(2018) 
to_year = int(2022)

# Inicializamos los parámetros de la solicitud
params = {'from': from_year, 'to': to_year}

# Establecemos los límites de resultados
max_resultados = 200
resultados_obtenidos = 0

# Lista para almacenar los resultados de las canciones
canciones_lista = []

# Bucle para obtener los resultados paginados
while resultados_obtenidos < max_resultados:
    
    params.update({
        'format': 'json', 
        'limit': 50,  # Número máximo de resultados por solicitud
        'page': (resultados_obtenidos // 50) + 1  # Página que vamos a consultar
    })
    
    # Realizamos la solicitud a la API
    llamada = requests.get(url, params=params)
    
        
        # Verificamos si la respuesta contiene las canciones
    if 'tracks' in llamada.json():
        #tracks = top_canciones_fecha['tracks']['track']
            
        # Añadimos las canciones a la lista
        for track in llamada.json()['tracks']['track']:
                
            # Creamos un diccionario con los datos de la canción
            cancion_top= {
                'name': track ['name'],
                'artist': track ['artist']['name']
                }
                
            # Añadimos la canción a la lista
            canciones_lista.append(cancion_top)
            
        # Actualizamos el número de resultados obtenidos
        resultados_obtenidos += len (llamada.json()['tracks']['track'])
```
